```python
weather = pd.read_csv('results/weather_airport_fixed.csv', low_memory=False)
#airports = pd.read_csv('results/backups/ABI_weather_backup.csv')
```

```python
flights = pd.read_pickle('data/flights_fixed.pkl')
```

```python
#fl1 = flights[:100]
fl1 = flights
```

```python
fl1['arrival_datetime'] = fl1.apply(lambda x: x['arrival_datetime'].round(freq='H'), axis=1)
fl1['departure_datetime'] = fl1.apply(lambda x: x['departure_datetime'].round(freq='H'), axis=1)
```

```python
fl1['arrival_datetime'] = fl1.apply(lambda x: x['arrival_datetime'].tz_localize(None), axis=1)
fl1['departure_datetime'] = fl1.apply(lambda x: x['departure_datetime'].tz_localize(None), axis=1)
```

```python
pd.set_option("display.max_columns", None)
display(fl1)
```

```python
#fl1['departure_datetime'] = fl1.apply(lambda x: x['departure_datetime'].astype(Int64), axis=1)
fl1['departure_datetime'] = fl1['departure_datetime'].astype(int)
```

```python
fl1['arrival_datetime'] = fl1['arrival_datetime'].astype(int)
```

```python
wh1 = weather
```

```python
wh1['Date time'] = wh1.apply(lambda x: pd.to_datetime(x['Date time']), axis=1)
```

```python
wh1['Date time'] = wh1['Date time'].astype(int)
```

```python
wh1 = wh1.drop(columns=["Name","Resolved Address","Longitude","Latitude","Maximum Temperature","Minimum Temperature","Address","Precipitation Cover",
                       "Heat Index","Wind Chill","Info"],axis=1)
```

```python
#wh1['DataFrame Column'] = wh1['DataFrame Column'].fillna(0)
```

```python
#merge departure
res = pd.merge(fl1, wh1, left_on=['ORIGIN_AIRPORT','departure_datetime'],  right_on = ['airport_code','Date time'])
```

```python
res = res.drop(columns=["Date time","airport_code","departure_datetime"],axis=1)
res = res.rename(columns={"Visibility":"Visibility_dep","Temperature":"Temperature_dep",
                         "Dew Point":"Dew_Point_dep","Relative Humidity":"Relative_Humidity_dep",
                         "Wind Speed":"Wind_Speed_dep","Wind Gust":"Wind_Gust_dep",
                         "Wind Direction":"WindDirection_dep",
                         "Precipitation":"Precipitation_dep","Snow Depth":"Snow_Depth_dep",
                         "Cloud Cover":"Cloud_Cover_dep","Sea Level Pressure":"Sea_Lv_Pressure_dep",
                         "Conditions":"Conditions_dep","Weather Type":"Weather_Type_dep"})
```

```python
#merge arrival
res = pd.merge(res, wh1, left_on=['DESTINATION_AIRPORT','arrival_datetime'],  right_on = ['airport_code','Date time'])
```

```python
res = res.drop(columns=["Date time","airport_code","arrival_datetime"],axis=1)
res = res.rename(columns={"Visibility":"Visibility_arv","Temperature":"Temperature_arv",
                         "Dew Point":"Dew_Point_arv","Relative Humidity":"Relative_Humidity_arv",
                         "Wind Speed":"Wind_Speed_arv","Wind Gust":"Wind_Gust_arv",
                         "Wind Direction":"WindDirection_arv",
                         "Precipitation":"Precipitation_arv","Snow Depth":"Snow_Depth_arv",
                         "Cloud Cover":"Cloud_Cover_arv","Sea Level Pressure":"Sea_Lv_Pressure_arv",
                         "Conditions":"Conditions_arv","Weather Type":"Weather_Type_arv"})
```
